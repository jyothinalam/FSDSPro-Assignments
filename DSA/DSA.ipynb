```python
class Node:
    def __init__(self, data):
        self.data = data
        self.next = None

class LinkedList:
    def __init__(self, *data):

        self.head = None

        # Create linked list if data is passed
        for i in data:
            new_node = Node(i)
            if self.head == None:
                self.head = new_node
            else:
                current_node = self.head
                while current_node.next:
                    current_node = current_node.next
                current_node.next = new_node
        return

    def traverse_linkedlist(self, *list):

        if len(list) == 0:
            current = self.head
        else:
            current = list[0]

        while current:
            if current.next is None:
                print(current.data, end=" ")
            else:
                print(current.data,"->", end=" ")
            current = current.next
        
    def rev_linkedlist(self):

        current = self.head
        previous = None # Stores revered linked list

        while current is not None:

            next_node = current.next

            current.next = previous
            previous = current

            current = next_node

        # Return the head of reversed linked list
        return previous
    
    def merge_sorted_lists(self, list1, list2):
  
        # Create a dummy node to simplify 
        # the merging process
        dummy = Node(-1)
        curr = dummy

        head1 = list1.head
        head2 = list2.head

        # Iterate through both linked lists
        while head1 is not None and head2 is not None:
            #print(head1.data, head2.data)
            
            # Add the smaller node to the merged list
            if head1.data <= head2.data:
                curr.next = head1
                head1 = head1.next
            else:
                curr.next = head2
                head2 = head2.next
            curr = curr.next

        # If any list is left, append it to the merged list
        if head1 is not None:
            curr.next = head1
        else:
            curr.next = head2

        # Return the merged list starting from 
        # the next of dummy node
        self.head = dummy.next
        return self.head
    
    def list_length(self):

        temp = self.head
        count = 0
        
        while(temp != None):
            count += 1
            temp = temp.next
            
        return count

    def remove_nth_node(self, n):

        length = self.list_length()

        node_from_beginning = length - n + 1

        prev = None
        temp = self.head

        for i in range(1, node_from_beginning):
            prev = temp
            temp = temp.next

        if(prev == None):
            self.head = self.head.next
        else:
            prev.next = prev.next.next

    def rearrange_even_odd(self):
         
        # Corner case 
        if (self.head == None): 
            return None
    
        # Initialize first nodes of 
        # even and odd lists 
        odd = self.head 
        even = self.head.next
    
        # Remember the first node of even list so 
        # that we can connect the even list at the 
        # end of odd list. 
        evenFirst = even 
    
        while (1 == 1): 
                
            # If there are no more nodes, 
            # then connect first node of even 
            # list to the last node of odd list 
            if (odd == None or even == None or (even.next) == None): 
                odd.next = evenFirst 
                break
    
            # Connecting odd nodes 
            odd.next = even.next
            odd = even.next
    
            # If there are NO more even nodes after current odd. 
            if (odd.next == None): 
                even.next = None
                odd.next = evenFirst 
                break
    
            # Connecting even nodes 
            even.next = odd.next
            even = odd.next
        return list.head
```

```python
# Problem 4: Find the intersection point of two linked lists.

# Input: List 1: 1 -> 2 -> 3 -> 4, List 2: 9 -> 8 -> 3 -> 4
# Output: Node with value 3

def get_intersection(diff, list1, list2):
    curr1 = list1.head
    curr2 = list2.head

    # Move the pointer forward by d nodes
    for _ in range(diff):
        if not curr1:
            return None
        curr1 = curr1.next

    # Move both pointers until they intersect
    while curr1 and curr2:
        if curr1.data == curr2.data:
            return curr1
        curr1 = curr1.next
        curr2 = curr2.next

    return None

# Function to get the intersection point of two linked lists
def intersect_point_lists(list1, list2):
  
    # Count the number of nodes in both linked lists
    len1 = list1.list_length()
    len2 = list2.list_length()

    diff = 0

    # If the first list is longer
    if len1 > len2:
        diff = len1 - len2
        return get_intersection(diff, list1, list2)
    else:
        diff = len2 - len1
        return get_intersection(diff, list2, list1)

## Driver code
list1 = LinkedList(1, 2, 3, 4)
list2 = LinkedList(9, 8, 3, 4)

print("\nLinked List1 :", end=' ')
list1.traverse_linkedlist()

print("\nLinked List2 :", end=' ')
list2.traverse_linkedlist()

intersection_Point = intersect_point_lists(list1, list2)

if not intersection_Point:
    print("No intersection point found")
else:
    print("\n\nIntersection point is node with value :", intersection_Point.data)
```

```python
# Problem 5: Remove duplicates from a sorted linked list.
# Input: 1 -> 1 -> 2 -> 3 -> 3
# Output: 1 -> 2 -> 3

def remove_duplicates(list):
    curr = list.head

    # Traverse the list
    while curr and curr.next:
      
        # Check if next value is the same as curr
        if curr.data == curr.next.data:
            next_next = curr.next.next
            curr.next = next_next
        else:
            curr = curr.next

    return list.head

## Driver code
list = LinkedList(1, 1, 2, 3, 3)

print("\nLinked List1 :", end=' ')
list.traverse_linkedlist()

duplicated = remove_duplicates(list)

print("\nLinked List after removing duplicates:", end=' ')
list.traverse_linkedlist(duplicated)
```

```python
# Problem 8: Reverse nodes in a linked list in groups of k
# Input: 1 -> 2 -> 3 -> 4 -> 5, k = 3
# Output: 3 -> 2 -> 1 -> 4 -> 5

def reverse_K_nodes(head, k):
    curr = head
    prev = None
    next = None
    count = 0

    while curr is not None and count < k:
        next = curr.next
        curr.next = prev
        prev = curr
        curr = next
        count += 1

    return prev

# Recursive function to reverse in groups of K
def reverse_K_group(head, k):
    if head is None:
        return head

    temp = head
    count = 0
    while temp is not None and count < k:
        temp = temp.next
        count += 1

    groupHead = reverse_K_nodes(head, k)
    
    # Recursion for the next group
    head.next = reverse_K_group(temp, k)

    return groupHead

## Driver code
list = LinkedList(1, 2, 3, 4, 5)
k = 3

print("\nOriginal Linked List :", end=' ')
list.traverse_linkedlist()

print(f"\nReversed Linked List in groups of {k},", end= ' ')
list.head = reverse_K_group(list.head, k)

list.traverse_linkedlist()
```

```python
# Problem 9: Determine if a linked list is a palindrome.

# Input: 1 -> 2 -> 2 -> 1
# Output: True

def is_palindrome_recursive(end, start):
    
    # base case
    if end is None:
        return True

    # Recursively check the right side.
    right = is_palindrome_recursive(end.next, start)

    # Compare the start and end nodes.
    ans = right and start[0].data == end.data

    # Update the start node
    start[0] = start[0].next

    return ans

# Function to check whether the list is palindrome
def is_palindrome(head):
  
    # Set starting node to head
    start = [head]

    # Recursively check the ll and return
    return is_palindrome_recursive(head, start)

## Driver code
list = LinkedList(1, 2, 2, 1)

print("\nOriginal Linked List :", end=' ')
list.traverse_linkedlist()

print("\nIs the linked list a palindrome? ", is_palindrome(list.head))
```

```python
# Problem 10: Rotate a linked list to the right by k places
# Input: 1 -> 2 -> 3 -> 4 -> 5, k = 2
# Output: 4 -> 5 -> 1 -> 2 -> 3

def right_rotate(head, k):

    # If the linked list is empty or no rotations are needed, 
    # return the original linked list
    if k == 0 or head is None:
        return head

    curr = head
    length = 1

    # Find the length of the linked list
    while curr.next is not None:
        curr = curr.next
        length += 1

    # Modulo k with length of linked list to handle
    # large values of k
    k %= length

    if k == 0:
        curr.next = None
        return head

    # Make the linked list circular
    curr.next = head

    # Traverse the linked list to find the kth node
    curr = head
    for _ in range(1, k):
        curr = curr.next

    # Update the (k + 1)th node as the new head
    newHead = curr.next

    # Break the loop by updating the next pointer
    # of kth node
    curr.next = None
    return newHead

## Driver code
list = LinkedList(1, 2, 3, 4, 5)
k = 2

print("\nOriginal Linked List :", end=' ')
list.traverse_linkedlist()

list.head = right_rotate(list.head, k)

print(f"\nLinked list after rotating to the right by {k} places : ", end=' ')
list.traverse_linkedlist()
```

```python
# Problem 13: Given a non-negative number represented as a linked list, add one to it

# Input: 1 -> 2 -> 3 (represents the number 123)
# Output: 1 -> 2 -> 4 (represents the number 124)

def add_with_carry(head):
    
    # If linked list is empty, return carry
    if head is None:
        return 1
    
    # Add carry returned by the next node call
    res = head.data + add_with_carry(head.next)
    
    # Update data and return new carry
    head.data = res % 10
    return res // 10

def add_one(head):
    
    # Add 1 to linked list from end to beginning
    carry = add_with_carry(head)
    
    # If there is carry after updating all nodes,
    # then we need to add a new node to the linked list
    if carry:
        newNode = Node(carry)
        newNode.next = head
        
        # New node becomes head now
        return newNode
    
    return head

## Driver code
list = LinkedList(1, 2, 3)

print("\nOriginal Linked List :", end=' ')
list.traverse_linkedlist()

list.head = add_one(list.head)

print("\nLinked List after adding 1:", end=' ')
list.traverse_linkedlist()
```

```python
# Problem 14: Given a sorted array and a target value, return the index if the target is found. If not, return the
# index where it would be inserted.

# Input: nums = [1, 3, 5, 6], target = 5
# Output: 2

def arr_find_index(arr, length, target):
 
    # Start and end indices of the array
    start = 0
    end = length - 1
 
    # Traverse through the array using binary search
    while start <= end:
 
        mid = (start + end)//2
 
        if arr[mid] == target:
            return mid
 
        elif arr[mid] < target:
            start = mid + 1
        else:
            end = mid - 1
 
    # Return insertion index
    return end + 1
 
## Driver Code
arr = [1, 3, 5, 6]
length = len(arr)
target = 5
print("Original Array: ", arr)
print()
print(f"Index at which {target} to be inserted in the array:", arr_find_index(arr, length, target))
```

```python
# Problem 15: Find the minimum element in a rotated sorted array.

# Input: [4, 5, 6, 7, 0, 1, 2]
# Output: 0

def arr_find_min(arr):
    start, end = 0, len(arr) - 1

    # Find minimum element using binary search
    while start < end:
      
        # The current subarray is already sorted, 
        # the minimum is at the low index 
        if arr[start] < arr[end]:
            return arr[start]
           
        # We reach here when we have at least
        # two elements and the current subarray
        # is rotated
        mid = (start + end) // 2

        # The right half is not sorted. So 
        # the minimum element must be in the
        # right half.
        if arr[mid] > arr[end]:
            start = mid + 1
          
        # The right half is sorted. Note that in 
        # this case, we do not change high to mid - 1
        # but keep it to mid. As the mid element
        # itself can be the smallest
        else:
            end = mid

    return arr[start]

## Driver Code
arr = [4, 5, 6, 7, 0, 1, 2]
print("Original Array: ", arr)
print(f"Minimum element in the Array:", arr_find_min(arr))
```

```python
# Problem 16: Search for a target value in a rotated sorted array.

# Input: nums = [4, 5, 6, 7, 0, 1, 2], target = 0
# Output: 4

def arr_rotated_search(arr, target):
  
    start = 0
    end = len(arr) - 1

    # Search for a target element in a sorted array using binary search
    while start <= end:
        mid = start + (end - start) // 2

        # If key found, return the index
        if arr[mid] == target:
            return mid

        # Left half is sorted
        if arr[mid] >= arr[start]:
          
            # Target lies within this sorted half, move end to mid - 1
            if target >= arr[start] and target < arr[mid]:
                end = mid - 1
              
            # Otherwise, move start to mid + 1
            else:
                start = mid + 1
          
        # Right half is sorted
        else:
          
            # Target lies within this sorted half, move start to mid + 1
            if target > arr[mid] and target <= arr[end]:
                start = mid + 1
              
            # Otherwise, move end to mid - 1
            else:
                end = mid - 1

    # Target not found
    return -1

## Driver Code
arr = [4, 5, 6, 7, 0, 1, 2]
target = 0
print("Original Array: ", arr)

found_index = arr_rotated_search(arr, target)

if (found_index < 0):
    print(f"Target element {target} is not found")
else:
    print(f"Target element {target} is found at the index:", found_index)
```

```python
# Problem 18: Given a m x n matrix where each row and column is sorted in ascending order, count the number
# of negative numbers.

# Input: grid = [[4, 3, 2, -1], [3, 2, 1, -1], [1, 1, -1, -2], [-1, -1, -2, -3]]
# Output: 8

def matrix_count_negatives(grid):
    rows, columns = len(grid), len(grid[0])
    i, j = rows - 1, 0
    ans = 0
    
    while i >= 0 and j < columns:
        if grid[i][j] < 0:
            ans += columns - j
            i -= 1
        else:
            j += 1
    return ans

def matrix_display(grid):
    
    rows, columns = len(grid), len(grid[0])
    
    print("\nElements of the Matrix:")

    for i in range(rows):
        for j in range(columns):
            print(grid[i][j], end=" ")
        print()

## Driver code

grid = [
    [4, 3, 2, -1], 
    [3, 2, 1, -1], 
    [1, 1, -1, -2], 
    [-1, -1, -2, -3]
    ]

print("Elements in the grid: ", grid)
print("Number of negative elements in the grid: ", matrix_count_negatives(grid))
```

```python
# Problem 19: Given a 2D matrix sorted in ascending order in each row, and the first integer of each row is
# greater than the last integer of the previous row, determine if a target value is present in the matrix.

# Input: matrix = [[1, 3, 5, 7], [10, 11, 16, 20], [23, 30, 34, 60]], target = 3
# Output: True

def matrix_search(matrix, target):
    rows = len(matrix)
    columns = len(matrix[0])

    start, end = 0, rows * columns - 1
    while start <= end:
        mid = (start + end) // 2

        # Find row and column of element at mid index
        row = mid // rows
        col = mid % columns

        # If target is found, return true
        if matrix[row][col] == target:
            return True

        # If target is greater than matrix[row][col], search in
        # right half
        if matrix[row][col] < target:
            start = mid + 1

        # If x is less than mat[row][col], search in
        # left half
        else:
            end = mid - 1

    return False

def matrix_display(grid):
    
    rows, columns = len(grid), len(grid[0])
    
    print("\nElements of the Matrix:")

    for i in range(rows):
        for j in range(columns):
            print(grid[i][j], end=" ")
        print()

## Driver code
matrix = [
    [1, 3, 5, 7], 
    [10, 11, 16, 20], 
    [23, 30, 34, 60]
    ]

target = 3

print("Elements in the Matrix: ", matrix)

if (matrix_search(matrix, target)):
    print("Element",target,"found in the matrix")
else:
    print("Element",target," not found in the matrix")
```

```python
# Problem 20: Find Median in Two Sorted Arrays
# Problem: Given two sorted arrays, find the median of the combined sorted array.

# Input: nums1 = [1, 3], nums2 = [2]
# Output: 2.0

def array_merge(left, right):
    result = []
    i = j = 0

    while (i < len(left)) and (j < len(right)):
        if left[i] < right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    result.extend(left[i:])
    result.extend(right[j:])

    return result

## Driver code
num1 = [1, 3]
num2 = [2]

print("Sorted Arrays: ", num1, num2)

combined_array = array_merge(num1, num2)
length = len(combined_array)

print("Merged sorted array: ", combined_array)

mid_index = length // 2

if length % 2 == 1:
    median = combined_array[mid_index]
else:
    median = float(combined_array[mid_index] + combined_array[mid_index + 1]) / 2

print("Median of the two sorted arrays: ", median)
```

```python
# Problem 21: Given a sorted character array and a target letter, find the smallest letter in the array that is
# greater than the target.

# Input: letters = ['c', 'f', 'j'], target = a
# Output: 'c

def chararr_next_greatest_alphabet(letters, target):
 
    length = len(letters)

    if(target >= letters[length-1]):
       return letters[0]
    
    start = 0
    end = len(letters) - 1
    ans = -1
    
    while (start <= end):
 
        mid = int((start + end) / 2)
 
        if (letters[mid] > target):
            end = mid - 1
            ans = mid
        else:
            start = mid + 1
 
    # Return the smallest element
    if (letters[ans] < target):
        return letters[0]
    else:
        return letters[ans]

## Driver code
letters = ['c', 'f', 'j']
target = 'a'

print("Sorted Arrays: ", letters)

result = chararr_next_greatest_alphabet(letters, target)

print(f"Smallest letter greater than the '{target}' is '{result}'")
```

```python
# Problem 22: Given an array with n objects colored red, white, or blue, sort them in-place so that objects of
# the same color are adjacent, with the colors in the order red, white, and blue.

# Input: nums = [2, 0, 2, 1, 1, 0]
# Output: [0, 0, 1, 1, 2, 2]

def sort_012(arr):
    cnt_0 = 0
    cnt_1 = 0
    cnt_2 = 0

    # Count 0s, 1s and 2s
    for num in arr:
        if num == 0:
            cnt_0 += 1
        elif num == 1:
            cnt_1 += 1
        else:
            cnt_2 += 1

    idx = 0
    
    # Place all the 0s
    for i in range(cnt_0):
        arr[idx] = 0
        idx += 1

    # Place all the 1s
    for i in range(cnt_1):
        arr[idx] = 1
        idx += 1

    # Place all the 2s
    for i in range(cnt_2):
        arr[idx] = 2
        idx += 1

## Driver code
nums = [2, 0, 2, 1, 1, 0]

print("Original Array: ", nums)
sort_012(nums)
print("After Sorting Array: ", nums)
```

```python
# Problem 23: Find the kth largest element in an unsorted array.

# Input: nums = [3, 2, 1, 5, 6, 4], k = 2
# Output: 5

# Funtion to sort the list
def arr_sort(num_list, length):
    for i in range(0, length):
        for j in range(i+1, length):
            if num_list[i] >= num_list[j]:
                num_list[i], num_list[j] = num_list[j], num_list[i]

## Driver code
nums = [3, 2, 1, 5, 6, 4]
length = len(nums)
k = 2

print("Original Array: ", nums)
arr_sort(nums, length)
print("Sorted Array: ", nums)

print(f"kth largest element of nums for k = {k} is {nums[length-k]}")
```

```python
# Problem 24: Given an unsorted array, reorder it in-place such that nums[0] <= nums[1] >= nums[2] <= nums[3]...
# Input: nums = [3, 5, 2, 1, 6, 4]
# Output: [3, 5, 1, 6, 2, 4]

def arr_wiggle_sort(nums):
    arr = sorted(nums)
    length = len(arr)
    
    i, j = (length - 1) >> 1, length - 1
    for k in range(length):
        if k % 2 == 0:
            nums[k] = arr[i]
            i -= 1
        else:
            nums[k] = arr[j]
            j -= 1

## Driver code
nums = [3, 2, 1, 5, 6, 4]

print("Original Array: ", nums)
arr_wiggle_sort(nums)
print("Wiggle Sorted Array: ", nums)
```

```python
# Problem 27: Implement linear search to find the index of a target element in an array.
# Input: [5, 3, 8, 2, 7, 4], target = 8
# Output: 2

def linear_search(arr, length, target):
    
    for i in range(length-1):
        if arr[i] == target:
            return i
    return -1

## Driver code
arr = [5, 3, 8, 2, 7, 4]
target = 8

length = len(arr)

print("Array: ", arr)

index = linear_search(arr, length, target)

if index == -1:
    print("Element not found")
else:
    print(f"Target element found at the index ", index)
```

```python
# Problem 28 Calculate the factorial of a given number.
# Input: 5
# Output: 120 (as 5! = 5 * 4 * 3 * 2 * 1 = 120)

def factorial(num):
    if num <= 1:
        return 1
    else:
        return (num * factorial(num-1))

num = 5
fact = factorial(num)

print(f"Factorial of {num}! = {fact}")
```

```python
# Problem 29: Check if a given number is a prime number.
# Input: 7
# Output: True

def is_prime(num):

    if num < 1:
        return False
    else:
        # Lood through 2 to num // 2
        for i in range(2, (num//2)+1):
            if (num % i) == 0:
                return False
        else:
            return True

## Driver code
num = 7

if is_prime(num) == True:
    print("True,",num, "is a prime number")
else:
    print(num, "is not a prime number")
```

```python
# Problem 30: Generate the Fibonacci series up to a given number n
# Input: 8
# Output: [0, 1, 1, 2, 3, 5, 8, 13]

def fibo(num):
    if num <= 1:
        return num
    
    return (fibo(num-1) + fibo(num-2))

## Driver code    
num = 8

arr = []
for i in range(num):
    arr.append(fibo(i))

print(num,"Fibonacci series:",arr)
```

```python
# Problem 31: Calculate the power of a number using recursion.
# Input: base = 3, exponent = 4
# Output: 81 (as 3^4 = 3 * 3 * 3 * 3 = 81)

def power(base, exponent):
    if base <= 1 or exponent <= 0:
        return 1
    
    return base * power(base, exponent-1)

base = 3
exponent = 4
print(f"{base}^{exponent} = ",power(base, exponent))
```
