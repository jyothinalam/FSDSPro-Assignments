Problem 1: Bank Account. Create a class representing a bank account with attributes like account number, account holder name, and balance. Implement methods to deposit and withdraw money from the account.

```python
print("Q1 Create a class representing a bank account with attributes like account number, account holder name, and balance. Implement methods to deposit and withdraw money from the account.\nAns.")

class bankaccount:
    
    def __init__(self,acc_number,acc_name,min_balance):
        self.acc_number = acc_number
        self.acc_name = acc_name
        self.balance = min_balance
        print(f"\t\tAccount {self.acc_number} created successfully.")
    
    def deposit(self, acc_number, acc_name, dep_amount):
        if (self.acc_number == acc_number) and (self.acc_name == acc_name):
            self.balance = self.balance + dep_amount
            print(f"\t\tAmount deposited successfully in {acc_number}. Your current balance: Rs.{self.balance}")
        else:
            print("\t\tAccount Number and/or Account Name is invalid. Unable to deposit the amount.")
                                                
    def withdraw(self, acc_number, acc_name, withdraw_amount):
        if (self.acc_number == acc_number) and (self.acc_name == acc_name):
            if self.balance < withdraw_amount:
                print("\t\tInsufficient Funds in the account {acc_number}")
            else:
                self.balance = self.balance - withdraw_amount
                print(f"\t\tAmount Rs.{withdraw_amount} withdrawn successfully from account {acc_number}. Your current balance: Rs.{self.balance}")
        else:
            print("\t\tAccount Number and/or Account Name is invalid. Unable to withdraw the amount.")
    
print("\tBank Account Creation.\n")
obj_acc_holder1 = bankaccount("ABC0001", "Jyothi", 5000)
obj_acc_holder2 = bankaccount("ABC0002", "Maccu", 5000)

print("\n\tAccount Transactions - Deposit and Withdrawal.\n")
obj_acc_holder1.deposit("ABC0001", "Jyothi", 1000)
obj_acc_holder1.withdraw("ABC0001", "Jyothi", 500)
print("\n")
obj_acc_holder2.deposit("ABC0002", "Maccu", 1000)
obj_acc_holder2.withdraw("ABC0001", "Maccu", 6000)
```

```python
print("Q4 Create classes representing a library and a book. Implement methods to add books to the library, borrow books, and display available books.\nAns.")

class library:
    
    books = []
    
    def add_book(self, subject, name, author):
        self.subject = subject
        self.name = name
        self.author = author
        self.status = "Available"
        library.books.append({'subject':self.subject,'name':self.name,'author':self.author,'status':self.status})
        print(f"\t  Book \'{name}\' added successfully.")
            
    def borrow_book(self, subject, name, author):
        if self.status == "Available":
            if (self.subject == subject) and (self.name == name) and (self.author == author):
                self.status = "Borrowed"
                for i in range(len(library.books)):
                    #print(book[i].subject)
                    if (library.books[i]['subject'] == subject) and (library.books[i]['name'] == name) and (library.books[i]['author'] == author):
                        library.books[i]['status'] = "Borrowed"
            else:
                print(f"\tBook by subject {subject}, name {name} and author {author} is not available.")
        else:
            print(f"\tBook by subject {subject}, name {name} and author {author} is already borrowed.")

def display_books(key):
    if key == 'all':
        for book in library.books:
            print(book)
    elif key == 'available':
        for i in range(len(library.books)):
            if library.books[i]['status'] == "Available":
                print(library.books[i])
    return

obj_book1 = library()
obj_book2 = library()
obj_book3 = library()
obj_book4 = library()
obj_book5 = library()

print("\n\t-------------------------------------")
print("\t   Add books to the library:")
print("\t-------------------------------------")
obj_book1.add_book("Data Science", "Full Stack Data Science", "DS Author1")
obj_book2.add_book("Data Science", "Data Science For Biginners", "DS Author2")
obj_book3.add_book("Data Science", "Data Science Project Management", "DS Author3")
obj_book4.add_book("DSA", "DSA For Beginners", "DSA Author1")
obj_book5.add_book("DSA", "DSA Complete Guide", "DSA Author2")

print("\n\t-------------------------------------")
print("\t   List of books in Library:")
print("\t-------------------------------------")
display_books('all')

obj_book2.borrow_book("Data Science", "Data Science For Biginners", "DS Author2")
obj_book5.borrow_book("DSA", "DSA Complete Guide", "DSA Author2")

print("\n\t-------------------------------------")
print("\t   Books availability in Library:")
print("\t-------------------------------------")
display_books('available')
```

Problem 5: Product Inventory Create classes representing a product and an inventory system. Implement methods to add products to the inventory, update product quantity, and display available products.

```python
print("Q5 Product Inventory. Create classes representing a product and an inventory system. Implement methods to add products to the inventory, update product quantity, and display available products.\nAns.")

class Inventory:
    
    inventory = []

    ## Define a constructor to intialize a product in inventory
    def __init__(self):
        pass
        
    ## Define a method to add a new inventory product
    def add_inventory(self, prod_name, prod_quantity):
        ## Append new inventory and create product ID automatically
        if self.inventory == []:
            self.inventory.append({'prod_ID':1, 'prod_name':prod_name, 'prod_quantity':prod_quantity})
        else:
            ## Get the current prod_ID from the inventory and create a new entry
            prod_ID = max([i['prod_ID'] for i in self.inventory])
            self.inventory.append({'prod_ID':prod_ID+1, 'prod_name':prod_name, 'prod_quantity':prod_quantity})
        print(f"\t{prod_name} is added in the inventory")
            
    ## Define a method to update an existing inventory product
    def update_inventory(self, prod_name, prod_quantity):
        for prod in self.inventory:
            if prod['prod_name'] == prod_name:
                prod['prod_quantity'] += prod_quantity
                print(f"\t{prod_name} is updated with the new inventory")
                return

    ## Define a method to generate a full list of inventory
    def list_inventory(self):
        for prod in self.inventory:
            print("\t",prod)

## Driver code
print("---------------------------------------------")
print("\t\tInventory Creation")
print("---------------------------------------------")
obj_prod1 = Inventory()
obj_prod2 = Inventory()
obj_prod3 = Inventory()
obj_prod4 = Inventory()

print("\tAdd Inventory")
print("     -------------------")
obj_prod1.add_inventory("Apples",100)
obj_prod2.add_inventory("Banana",200)
obj_prod3.add_inventory("Carrot",50)
obj_prod4.add_inventory("Capsicum",20)

print("\n")
print("\tAvailable stock of Inventory")
print("     -----------------------------------")
obj_prod1.list_inventory()

print("\n")
print("\tUpdate Inventory")
print("     -----------------------------------")
obj_prod1.update_inventory("Apples",20)
obj_prod2.update_inventory("Banana",50)
obj_prod3.update_inventory("Carrot",20)
obj_prod4.update_inventory("Capsicum",10)

print("\n")
print("\tAvailable stock of Inventory")
print("     -----------------------------------")
obj_prod1.list_inventory() 
```

Problem 6: Shape Calculation Create a class representing a shape with attributes like length, width, and height. Implement methods to calculate the area and perimeter of the shape.

```python
print("Q6 Shape Calculation. Create a class representing a shape with attributes like length, width, and height. Implement methods to calculate the area and perimeter of the shape.\nAns.")

import math

## Define a parent class with area and perimeter methods
class shape:
    
    ## Define method calc_area. This method will be overridden by sub classes
    def calc_area(self):
        pass
    
    ## Define method calc_perimeter. This method will be overridden by sub classes
    def calc_perimeter(self):
        pass

## Define a child class circle 
class circle(shape):
    
    ## Define constructor
    def __init__(self, radius):
        self.radius = radius
    
    ## Define Area method using formula A = Pi * radius^2
    def calc_area(self,radius):
        return math.pi * radius * radius
    
    ## Define Perimeter method using formula P = 2 * Pi * radius
    def calc_perimeter(self, radius):
        return 2 * math.pi * self.radius
    
## Define a child class square
class square(shape):
    
    ## Define constructor
    def __init__(self, side):
        self.side = side
    
    ## Define Area method using formula A = side * side
    def calc_area(self, side):
        return self.side * self.side
    
    ## Define Perimeter method using formula P = 4 * side
    def calc_perimeter(self,side):
        return 4 * self.side
    
## Define a child class rectangle
class rectangle(shape):
    
    ## Define constructor
    def __init__(self, length, width):
        self.length = length
        self.width = width
    
    ## Define Area method using formula A = length * width
    def calc_area(self, length, width):
        return self.length * self.width
    
    ## Define Perimeter method using formula P = 2 * (length + width)
    def calc_perimeter(self, length, width):
        return 2 * (self.length + self.width)
    
## Define a child class triangle
class triangle(shape):
    
    ## Define constructor
    def __init__(self, base, height, side1, side2, side3):
        self.base = base
        self.height = height
        self.side1 = side1
        self.side2 = side2
        self.side3 = side3
    
    ## Define Area method using formula A = base * height
    def calc_area(self, base, height):
        return 0.5 * self.base * self.height
    
    ## Define Perimeter method using formula P = side1 + side2 + side3
    def calc_perimeter(self, side1, sid2, side3):
        return self.side1 + self.side2 + self.side3
    

## Driver Code
## Declare objects of each geometrical type

radius = 5
obj_circle = circle(radius)
area_circle = obj_circle.calc_area(radius)
perimeter_circle = obj_circle.calc_perimeter(radius)

print("\n\tShape - Circle")
print(f"\t\tArea of the circle with radius {radius} = {round(area_circle, 2)}")
print(f"\t\tPerimeter of the circle with radius {radius} = {round(perimeter_circle, 2)}")

length = 3
width = 5
obj_rectangle = rectangle(length, width)
area_rect = obj_rectangle.calc_area(length, width)
perimeter_rect = obj_rectangle.calc_perimeter(length, width)

print("\n\tShape - Rectangle")
print(f"\t\tArea of the rectangle with length {length} and width {width} = {area_rect}")
print(f"\t\tPerimeter of the rectangle with length {length} and width {width} = {perimeter_rect}")

side = 10
obj_square = square(side)
area_square = obj_square.calc_area(side)
perimeter_square = obj_square.calc_perimeter(side)

print("\n\tShape - Square")
print(f"\t\tArea of the square with side {side} = {area_square}")
print(f"\t\tPerimeter of the square with side {side} = {perimeter_square}")

base = 4
height = 7
side1 = 3
side2 = 4
side3 = 5

obj_triangle = triangle(base, height, side1, side2, side3)
area_triangle = obj_triangle.calc_area(base, height)
perimeter_triangle = obj_triangle.calc_perimeter(side1, side2, side3)

print("\n\tShape - Triangle")
print(f"\t\tArea of the triangle with base {base} and height = {height} = {area_triangle}")
print(f"\t\tPerimeter of the triangle with sides {side1}, {side2} and {side3} = {perimeter_triangle}")
```

Problem 7: Student Management Create a class representing a student with attributes like student ID, name, and grades. Implement methods to calculate the average grade and display student details.

```python
print("Q7 Student Management. Create a class representing a student with attributes like student ID, name, and grades. Implement methods to calculate the average grade and display student details.\nAns.")

class student:
    
    def __init__(self, **stud_info):
        #print(stud_info)
        self.stud_ID = stud_info['stud_ID']
        self.stud_name = stud_info['stud_name']
        self.stud_subject1_grade = stud_info['stud_subject1_grade']
        self.stud_subject2_grade = stud_info['stud_subject2_grade']
        self.stud_subject3_grade = stud_info['stud_subject3_grade']
        print(f"\t Student with ID {self.stud_ID} and {self.stud_name} is created successfully.")
                
    def disp_stud_details(self, stud_ID):
        if self.stud_ID == stud_ID:
            print(f"\n\tStudent details with ID: {self.stud_ID}")
            print(f"\n\tStudent Name: {self.stud_name}")
            print(f"\tStudent Subject1 Grade: {self.stud_subject1_grade}")
            print(f"\tStudent Subject2 Grade: {self.stud_subject2_grade}")
            print(f"\tStudent Subject3 Grade: {self.stud_subject3_grade}")
        else:
            print(f"\tStudent with ID {stud_ID} doesn't exist.")
            
    def calc_avg_grade(self, stud_ID):
        if self.stud_ID == stud_ID:
            total_subject_grades = self.stud_subject1_grade+self.stud_subject2_grade+self.stud_subject3_grade
            avg_grade = total_subject_grades/3
            print(f"\n\tAverage grade of student with studentID {stud_ID} is {avg_grade}.")
        else:
            print("\tStudent with ID {stud_ID} doesn't exist.")
        
## Driver code
## Declare student objects

print("---------------------------------------------")
print("\t\tStudent Creation:")
print("---------------------------------------------")
obj_stud1 = student(stud_ID=100, stud_name='Jyothi',stud_subject1_grade=70, stud_subject2_grade=80, stud_subject3_grade=85)
obj_stud2 = student(stud_ID=101, stud_name='Padma',stud_subject1_grade=85, stud_subject2_grade=70, stud_subject3_grade=95)
obj_stud3 = student(stud_ID=102, stud_name='Maccu',stud_subject1_grade=90, stud_subject2_grade=90, stud_subject3_grade=90)

print("\n")
print("---------------------------------------------")
print("  Display list of students and their grades:")
print("---------------------------------------------")
## Display all students and their info
obj_stud1.disp_stud_details(100)
obj_stud2.disp_stud_details(101)
obj_stud3.disp_stud_details(102)

print("\n")
print("---------------------------------------------")
print("Display average grade of a specified student:")
print("---------------------------------------------")
## Calculate Average grade for a specified Student
stud_ID = 102
avg_grade = obj_stud3.calc_avg_grade(stud_ID)
```

Problem 8: Email Management Create a class representing an email with attributes like sender, recipient, and subject. Implement methods to send an email and display email details.

```python
print("Q8 Email Management. Create a class representing an email with attributes like sender, recipient, and subject. Implement methods to send an email and display email details.\nAns.")

class email_mgmt:
    
    # Domain name for emails
    global email_domain
    email_domain   = "@test.local"
    
    ## Define constrcutor to initialize user information.
    def __init__(self, first_name, last_name):
        self.first_name = first_name
        self.last_name  = last_name
        self.user_name  = first_name[0]  + last_name
        self.address   = self.user_name + email_domain
        self.sender = self.address
        self.emails = []
        print(f"\t   User with emailID {self.address} is created successfully.")
        
    ## Define method to send emails.
    def send_email(self, recipient, subject, message):
        
        self.emails.append({'sender':self.sender,'recipient':recipient,'subject':subject,'message':message})
        
        email_content = ("   from: " + self.sender + "\n" +
                        "\t   to: " + recipient + "\n" +
                        "\t   subject: " + subject + "\n\n" +
                        "\t\t"+message)
        print(f"\t{email_content}")
        print("\n\t---- Email Sent! successfully ----\n")
       
    ## Define method to display details of all the emails received by a specific user.
    def disp_email_details(self):
        print(f"\t   Emails received by {self.emails[0]['recipient']}\n")
        for email in self.emails:
            #print(email)
            email_content = ("\tfrom: " + email['sender'] + "\n" +
                        "\t\tto: " + email['recipient'] + "\n" +
                        "\t\tsubject: " + email['subject'] + "\n\n" +
                        "\t\t"+email['message'])
            print(f"\t{email_content}\n\n\t   ----- End of email -----\n")
        
## Driver Code
## 1. Create Users with First Name and Last Name
print("-------------------------------------")
print("\tUser Creation:")
print("-------------------------------------")
obj_user1 = email_mgmt('Jyothi','Nalamaru')
obj_user2 = email_mgmt('Padma','Nalamaru')

## 2. Send emails 
print("\n-------------------------------------")
print("\tSending Emails:")
print("-------------------------------------")
obj_user1.send_email("PNalamaru@test.local","Test1 Email","Hey there!\n\t\tHow are you?")
obj_user2.send_email("JNalamaru@test.local","Response Email","Hey there!\n\t\tI am doing good!")
obj_user1.send_email("PNalamaru@test.local","Test2 Email","Hey there!\n\t\tGood to here you are fine.")

## 3. Display emails received for a specifief user
print("-------------------------------------")
print("\tDisplay List of Emails:")
print("-------------------------------------")
obj_user1.disp_email_details()
```

Problem 10: ToDo List Create a class representing a ToDo list with attributes like tasks and due dates. Implement methods to add tasks, mark tasks as completed, and display pending tasks.

```python
print("Q10 Create a class representing a ToDo list with attributes like tasks and due dates. Implement methods to add tasks, mark tasks as completed, and display pending tasks.\nAns.")

import datetime

class ToDo_List:
    def add_task(self, task_ID, task_name, task_duedate):
        self.task_ID = task_ID
        self.task_name = task_name
        self.task_duedate = task_duedate
        self.task_status = "Pending"
        print(f"\tTask with ID {self.task_ID} added successfully.")
            
    def display_task(self,task_ID):
        if (self.task_ID == task_ID):  
            print(f"\nTask ID = {task_ID}\nTask Name = {self.task_name}\nTask Due Date = {self.task_duedate}\nTask Status = {self.task_status}")
        else:
            print(f"\t\tInvalid Task ID {task_ID}.")
              
    def complete_task(self, task_ID):
        if (self.task_ID == task_ID):
            self.task_status = "Complete"
        else:
            print(f"\t\tInvalid Task ID {self.task_ID}.")

obj_task1 = ToDo_List()
obj_task2 = ToDo_List()
obj_task3 = ToDo_List()

print("\n-------------------------------------")
print("\tAdd tasks to To-Do List:")
print("-------------------------------------")
obj_task1.add_task(1, "Data Science Pro Assignment Loops", datetime.datetime(2023, 10, 5))
obj_task2.add_task(2, "Data Science Pro Assignment Python Basics", datetime.datetime(2023, 9, 15))
obj_task3.add_task(3, "Data Science Pro Assignment OOPS", datetime.datetime(2023, 11, 5))

print("\n------------------------------------------")
print("\tList of tasks in To-Do List:")
print("------------------------------------------")
obj_task1.display_task(1)
obj_task2.display_task(2)
obj_task3.display_task(3)

obj_task1.complete_task(1)

print("\n------------------------------------------")
print("\tStatus of tasks in To-Do List:")
print("------------------------------------------")
obj_task1.display_task(1)
obj_task2.display_task(2)
obj_task3.display_task(3)
```
