#### 24. A company produces three different versions of a product: Standard, Premium, and Deluxe. The company wants to determine if there is a significant difference in customer satisfaction scores among the three product versions. They conducted a survey and collected customer satisfaction scores for each version from a random sample of customers. Perform an ANOVA test to determine if there is a significant difference in customer satisfaction scores.

  Use the following data:
  ```python

  # Sample data: Customer satisfaction scores for each product version
  standard_scores = [80, 85, 90, 78, 88, 82, 92, 78, 85, 87]
  premium_scores = [90, 92, 88, 92, 95, 91, 96, 93, 89, 93]
  deluxe_scores = [95, 98, 92, 97, 96, 94, 98, 97, 92, 99]
  ```

````python

standard_scores = [80, 85, 90, 78, 88, 82, 92, 78, 85, 87]
premium_scores = [90, 92, 88, 92, 95, 91, 96, 93, 89, 93]
deluxe_scores = [95, 98, 92, 97, 96, 94, 98, 97, 92, 99]

sum_standard_scores = sum(standard_scores) 
sum_premium_scores = sum(premium_scores)
sum_deluxe_scores = sum(deluxe_scores)

print(f"Sum of each sample = {sum_standard_scores}, {sum_premium_scores}, {sum_deluxe_scores}")

total_scores = sum_standard_scores + sum_premium_scores + sum_deluxe_scores

print("\nTotal sum of all the samples = ", total_scores)

n = len(standard_scores) 
N = len(standard_scores) + len(premium_scores) + len(deluxe_scores)

df_between_sample = 3 - 1
df_within_sample = N - 3

f_critical = 3.35

print(f"\ndof between samples = {df_between_sample}, dof within sample = {df_within_sample}")

y_square = 0
for i in range(len(standard_scores)):
    y_square += standard_scores[i] ** 2

for i in range(len(premium_scores)):
    y_square += premium_scores[i] ** 2

for i in range(len(deluxe_scores)):
    y_square += deluxe_scores[i] ** 2

print("\nSum of squares of each individual data = ", y_square)

ss_between_sample = (sum_standard_scores ** 2 + sum_premium_scores ** 2 + sum_deluxe_scores **2)/n - (total_scores ** 2)/N

print("\nSum of square between the samples = ", round(ss_between_sample, 2))

ss_within_sample = y_square - (sum_standard_scores ** 2 + sum_premium_scores ** 2 + sum_deluxe_scores **2)/n

print("\nSum of square within the sample = ", round(ss_within_sample, 2))

f_stats = (ss_between_sample/df_between_sample) / (ss_within_sample/df_within_sample)


if f_stats < f_critical:
    print(f"\n{round(f_stats, 2)} < {f_critical}. Accept the H0: All sample means are the same")
else:
    print(f"\n{round(f_stats, 2)} < {f_critical} Fail to Accept / Reject the H0: All sample means are not the same")
````

#### Steps to perform One-way ANOVA test:

#### Step1:  (Framing Hypothesis)

        H0: All sample means are equal
        Ha: Atleast one sample mean is not equal
  
#### Step2: (Significance Level)

$\alpha = 0.05$

#### Step3: (Calculate F statistics)

  $F_{statistics}$ = Variance between the samples (Mean Square between the samples / Variance within the sample (Mean Square within the sample)

Sum of squares between the sample = $\sum \sum (a_{i})^2/n - T^2/N$

  $\sum (a_{i})^2 = a_{1}^2 + a_{2}^2 + a_{3}^2$

  $T^2 = (a_{1} + a_{2} + a_{3})^2$

  n = each sample size = 10
  
  N = Total size of the samples = 30

Sum of squares within the sample = $\sum y_{i}^2 - \sum \sum (a_{i})^2/n$

$F_{statistics} = $ sum_between_sample/df1 / sum_between_sample/df2                                 

#### Step4: (F critical)

df1 =  3 - 1 = 2
df2 = 30 - 3 = 27

$F_{critical}^{0.05}$ = 3.35 

#### Step5: Conclusion

$F_{statistics} < F_{critical}$ Accept the H0 hypothesis. Mean of all the samples is the same.

$F_{statistics} > F_{critical}$ Reject the H0 hypothesis. Mean of all the samples is not the same.
