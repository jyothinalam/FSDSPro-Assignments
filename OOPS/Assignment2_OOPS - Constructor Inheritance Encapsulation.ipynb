```python
#Q10 How can you have multiple constructors in a Python class? Explain with an example.

print("Q10. How can you have multiple constructors in a Python class? Explain with an example.\nAns.")

print('''\nAs Python doesn't allow to define multiple constructors, we can achieve the same by using Constructor overloading.
We can define constructor to have arbitrary arguments (*args) as the parameter. Based on the number of arguments passed,
we can perform different operations. This is demonstrated by the example below.

------------------------------------------------------\n''')

class Student:
 
    ## Constructor overloading based on args
    def __init__(self, *args):
 
        num_args = len(args)
        
        ## If args == 2, then consider them as first name and last name
        if num_args == 2:
            self.firstname = args[0]
            self.lastname = args[1]
            print("Arguments received in Constructor: 2. First Name and Last Name.")
        
        ## If args > 2, then consider them as the list of elective subects
        elif num_args > 2:
            self.subjects = []
            for subject in args:
                self.subjects.append(subject)
            print("Arguments received in Constructor: > 2. List of Elective Subjects.")

    def show(self, key):
        if key == "name":
            print(f"   Name of the student is: {self.firstname} {self.lastname}:")
        elif key == "subjects":
            print(f"   Available Elective Subjects :", self.subjects)
                
## Driver Code
s1 = Student("Jyothi", "Nalamaru")
elective_subj = Student("English","Hindi","Science")

print("\nDisplay Student Full Name and the Available Elective Subjects: ")
s1.show("name")
elective_subj.show("subjects")
```

```python
#Q12 Explain the use of the `super()` function in Python constructors. Provide an example.

print("Q12. Explain the use of the `super()` function in Python constructors with an example.\nAns.")

print('''\nsuper() function is used to call a superclass Constructor in inheritance. When a subclass is created, its 
constructor must call the constructor of its parent class. This is done using the super() keyword, which calls the 
constructor of the parent class.

For Example - ''')

class Vehicle:
    # Constructor of Vehicle
    def __init__(self, engine):
        print('  Inside Parent Vehicle Constructor')
        self.engine = engine

class Car(Vehicle):
    # Constructor of Car
    def __init__(self, engine, max_speed):
        super().__init__(engine)
        print('  Inside Parent Car Constructor')
        self.max_speed = max_speed

class Electric_Car(Car):
    # Constructor of Electric Car
    def __init__(self, engine, max_speed, km_range):
        super().__init__(engine, max_speed)
        print('  Inside Child Electric Car Constructor')
        self.km_range = km_range

# Object of electric car
ev = Electric_Car('1500cc', 240, 750)
print(f"\n  All attributes are accessed from Child class :")
print(f'\tEngine={ev.engine}, Max Speed={ev.max_speed}, KM Range={ev.km_range}')
```

```python
#Q18. What is the "diamond problem" in multiple inheritance, and how does Python address it?

print("Q18. What is the \"diamond problem\" in multiple inheritance, and how does Python address it?\nAns.")

print('''\nIn inheritance, the diamond problem occurs when two classes (class B & class C) have a common parent class (A), and 
another class (D) has both those classes (class B & class C) as base classes.

                class A
                  / \\
                 /   \\
                /     \\
           class B    Class C
                \     /
                 \   /
                  \ / 
                Class D
                
In this scenario of inheritance, if class B and class C override (any) same method of class A and instance of class D calls 
that method, it becomes ambiguous to programming language, whether to call the overridden method from class B or the one from 
class C.

The diamond probelm in Python is resolved by using the concept called Method Resolution Order (MRO). This gives the order
in which the method is searched in the class hierarchy. Python interpreter uses MRO and executes class B method.

Demo of the diamond problem resolution -''')

class A:
    def display(self):
        print("  This method is from class A")
        
class B(A):
    def display(self):
        print("  This method is from class B")
        
class C(A):
    def display(self):
        print("  This method is from class C")
        
class D(B, C):
    pass

obj = D()
obj.display()
```

```python
#Q10. How can you access private attributes in Python? Provide an example demonstrating the use of name mangling.

print("Q10. How can you access private attributes in Python? Provide an example demonstrating the use of name mangling.\nAns\n")

print("Private attributes can be accessed using public method of the class as demonstrated below.")

class Student:
    
    # constructor
    def __init__(self, name, roll, branch): 
        
        # Initialize private members
        self.__name = name
        self.__roll = roll
        self.__branch = branch
 
    # public member function to access private attributes
    def display_details(self): 
            
        print("\nAccessing private attributes from public method of the class:")
        print("Name: ", self.__name)
        print("Roll: ", self.__roll)
        print("Branch: ", self.__branch)

## creating object    
st = Student("Jyothi", 1706256, "Data Science")
 
# calling public member function of the class
st.display_details()

# The below statement throws an error as we cannot access private attribute directly.
#print("Name :",st.__name) # ----- Cannor access the variable directly.

print('''\nAs we cannot access private attributes directly from outside the class, Python provides a process in which any given identifier 
with one trailing underscore and two leading underscores is textually replaced with the __ClassName__Identifier is known as 
Name mangling. In __ClassName__Identifier name, ClassName is the name of current class where identifier is present.

This can be demonstrated as below.''')

#Accessing private attribute using name mandling.
print("\nAccessing private attribute using name mandling: ")
print("Name :", st._Student__name)
print("Roll :", st._Student__roll)
print("Branch :", st._Student__branch)
```

```python
#Q13. What is data hiding, and why is it important in encapsulation? Provide examples.

print("Q13. What is data hiding, and why is it important in encapsulation? Provide examples.\nAns.\n")
      
print('''Data hiding is a part of object-oriented programming, which is generally used to hide the data information from the user.
The main working of data hiding is that it combines the data and functions into a single unit to conceal data within a class. 

It is important in encapsulation as it includes internal object details such as data members, internal working. Thus, maintaining
the data integrity and restricted access to the class member. To access encapsulated date, we declare class members as private 
so that no other class can access these data members. It is accessible only within the class.

We use getter() and setter() functions to access private attributes of the class as demonstrated in the below example.
''')

class Student:
    def __init__(self, name, age):
        # private member
        self.name = name
        self.__age = age

    # getter method
    def get_age(self):
        return self.__age

    # setter method
    def set_age(self, age):
        self.__age = age

stud = Student('Jyothi', 30)

print("Retrieving age using getter function:")
print('Name:', stud.name, 'and Age', stud.get_age())

print("\nChanging age using setter and then retrieving age using getter:")
stud.set_age(40)
# retrieving age using getter
print('Name:', stud.name, 'and Age', stud.get_age())
```

```python
#Q14. Create a Python class called `Employee` with private attributes for salary (`__salary`) and employee ID (`__employee_id`). 
#Provide a method to calculate yearly bonuses.

print("Q14. Create Employee class with private attributes salary and employee ID. Provide a method to calculate yearly bonus.\nAns.")
      
class Employee:
    
    def __init__(self, employee_id, salary):
        self.__employee_id = employee_id
        self.__salary = salary
        
        print(f"\t\tEmployee {self.__employee_id} created successfully.")
        
    def calc_annual_bonus(self):
        if self.__salary <= 5000:
            bonus = 10/100 * self.__salary
        elif self.__salary <= 10000:
            bonus = 20/100 * self.__salary
        else:
            bonus = 30/100 * self.__salary
        print(f"\t\tYearly bonus for Employee {self.__employee_id} is: {bonus}")
              
print("\tEmployee Creation.")
emp1 = Employee(101, 14000)
emp2 = Employee(102, 6000)
emp3 = Employee(103, 5000)

print("\n\tEmployee Bonus Calculation based on Salary:")
emp1.calc_annual_bonus()
emp2.calc_annual_bonus()
emp3.calc_annual_bonus()
```
