```python
#Q3. Create a Python class hierarchy for shapes (e.g., circle, square, triangle) and demonstrate polymorphism through a common
#method, such as `calculate_area()`.

print("Q3. Create a Python class hierarchy for shapes and demonstrate polymorphism through a common method, such as `calculate_area()`\nAns.\n")

class Shape:
    def calculate_area(self):
        raise NotImplementedError("Subclass must implement abstract method")

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def calculate_area(self):
        print("Area of a Circle", end=" ")
        return 3.14* (self.radius**2)

class Square(Shape):
    def __init__(self, side):
        self.side = side

    def calculate_area(self):
        print("Area of a Square",  end=" ")
        return self.side ** 2
    
class Triangle(Shape):
    def __init__(self, base, height):
        self.base = base
        self.height = height

    def calculate_area(self):
        print("Area of a Triangle", end=" ")
        return 0.5*self.base*self.height

shapes = [Circle(5), Square(10), Triangle(10, 5)]
for shape in shapes:
    print(f"from {type(shape).__name__} Class, {shape.calculate_area()}")

# Output:
# Drawing a circle
# Drawing a square
```

```python
#Q4. Explain the concept of method overriding in polymorphism. Provide an example.

print("Q4. Explain the concept of method overriding in polymorphism. Provide an example.\nAns.\n")

print('''Method overriding is an ability of polymorphism that allows a subclass or child class to provide a specific implementation of a method 
that is already provided by one of its super-classes or parent classes.
''')

class Parent(): 
    
    # Constructor 
    def __init__(self): 
        self.value = "Inside Parent"
        
    # Parent's show method 
    def show(self): 
        print(self.value) 
        
# Defining child class 
class Child(Parent): 
    
    # Constructor 
    def __init__(self): 
        super().__init__()  # Call parent constructor
        self.value = "Inside Child"
        
    # Child's show method 
    def show(self): 
        print(self.value) 
        
# Driver's code 
obj1 = Parent() 
obj2 = Child() 

obj1.show()  # Prints "Inside Parent"
obj2.show()  # Prints "Inside Child"
```

```python
#Q7. Discuss the use of abstract methods and classes in achieving polymorphism in Python. Provide an example using the `abc` module.

print("Q7. Discuss the use of abstract methods and classes in achieving polymorphism in Python. Provide an example using the `abc` module.\nAns.\n")

print('''Abstract methods and classes define a common interface for a group of subclasses. They cannot be instantiated themselves and require 
subclasses to provide implementations for abstract methods. Abstract Base Classes ensure that derived classes adhere to a specific protocol, thus 
supporting polymorphism. \n\nBelow is an example using 'abc' module:-\n''')

from abc import ABC, abstractmethod

class Animal(ABC):
    @abstractmethod
    def sound(self):
        pass

class Dog(Animal):
    def sound(self):
        return "Bark"

class Cat(Animal):
    def sound(self):
        return "Meow"

animals = [Dog(), Cat()]
for animal in animals:
    print(f"From class {type(animal).__name__}, Sound is - {animal.sound()}")
```

```python
#Q14. Create a Python class hierarchy for a banking system with various account types (e.g., savings, checking, credit card) 
#and demonstrate polymorphism by implementing a common `withdraw()` method.

print('''Q14. Create a Python class hierarchy for a banking system with various account types (e.g., savings, checking, credit card) and 
demonstrate polymorphism by implementing a common `withdraw()` method.\nAns.\n''')

class Account:
    def __init__(self, account_number, balance):
        self.account_number = account_number
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount

    def withdraw(self, account_number, amount):
        if self.account_number == account_number:
            if amount > self.balance:
                print("Insufficient funds.")
            else:
                self.balance -= amount
                print(f"Account Type: {self.account_type}, Amount withdrawn: {amount}, Remaining balance: {self.balance}")
        else:
            print("Invalid Account")

    def display_balance(self):
        print(f"Account Number: {self.account_number}, Account Type : {self.account_type}, Balance: {self.balance}")

class SavingsAccount(Account):
    def __init__(self, account_number, balance):
        super().__init__(account_number, balance)
        self.account_type = "Savings"
        
class CreditAccount(Account):
    def __init__(self, account_number, balance):
        super().__init__(account_number, balance)
        self.account_type = "Credit"

class CheckingAccount(Account):
    def __init__(self, account_number, balance):
        super().__init__(account_number, balance)
        self.account_type = "Checking"

class Bank:
    def __init__(self):
        self.accounts = []

    def add_account(self, account):
        self.accounts.append(account)

    def display_accounts(self):
        for account in self.accounts:
            account.display_balance()

bank = Bank()
sa_acct = SavingsAccount("12345", 10000)
ca_acct = CreditAccount("67890", 100000)
ch_acct = CheckingAccount("52344", 25000)

print("List of accounts created :\n")
bank.add_account(sa_acct)
bank.add_account(ca_acct)
bank.add_account(ch_acct)
bank.display_accounts()
print("\nWithdraw money using common method using polymorphism:\n")

sa_acct.withdraw("12345", 1000)
ca_acct.withdraw("67890", 1000)
ch_acct.withdraw("52344", 1000)
```

```python
#Q3. Create a Python class called `Shape` with an abstract method `calculate_area()`. Then, create child classes (e.g.,`Circle`,
#`Rectangle`) that implement the `calculate_area()` method. Provide an example of using these classes.

print('''Q3. Create a Python class called `Shape` with an abstract method `calculate_area()`. Then, create child classes (e.g.,`Circle`,
`Rectangle`) that implement the `calculate_area()` method. Provide an example of using these classes.\nAns.\n''')

from abc import ABC, abstractmethod

class Shape(ABC):

    @abstractmethod
    def calculate_area(self):
        pass  # This is an abstract method, no implementation here.

class Circle(Shape):
    def __init__(self, radius):
        self.radius = radius

    def calculate_area(self):
        print("Area of a Circle", end=" ")
        return 3.14* (self.radius**2)

class Rectangle(Shape):
    def __init__(self, length, width):
        self.length = length
        self.width = width

    def calculate_area(self):
        print("Area of a Rectangle",  end=" ")
        return self.length * self.width
    
shapes = [Circle(5), Rectangle(8, 6)]
for shape in shapes:
    print(f"from {type(shape).__name__} Class, {shape.calculate_area()}")
```

```python
#Q14. Create a Python class hierarchy for a restaurant system, using composition to represent menus, dishes, and ingredients.

print("Create a Python class hierarchy for a restaurant system, using composition to represent menus, dishes, and ingredients.\nAns.\n")

class Ingredient:
    def __init__(self, name, quantity, unit):
        self.name = name
        self.quantity = quantity
        self.unit = unit

    def __str__(self):
        return f"{self.quantity} {self.unit} of {self.name}"

class Dish:
    def __init__(self, name, description, price, ingredients):
        self.name = name
        self.description = description
        self.price = price
        self.ingredients = ingredients

    def __str__(self):
        return f"{self.name} - {self.description} (Rs. {self.price:.2f})"

class Menu:
    def __init__(self, name, dishes):
        self.name = name
        self.dishes = dishes

    def __str__(self):
        menu_str = f"\t\t\t\t{self.name}\n\n"
        for dish in self.dishes:
            menu_str += f"{dish}\n"
        return menu_str

class Restaurant:
    def __init__(self, name, menus):
        self.name = name
        self.menus = menus

    def display_menus(self):
        print(f"\t\t\tWelcome to {self.name}!")
        for menu in self.menus:
            print(menu)

## Driver code
ingredient1 = Ingredient(name="Tomato", quantity=2, unit="pieces")
ingredient2 = Ingredient(name="Cheese", quantity=150, unit="grams")
ingredient3 = Ingredient(name="Bread", quantity=2, unit="slices")

dish1 = Dish(name="Pizza", description="Classic pizza with tomato and cheese", price=400, ingredients=[ingredient1, ingredient2])
dish2 = Dish(name="Sandwich", description="Toasted sandwich with melted cheese", price=150, ingredients=[ingredient2, ingredient3])
dish3 = Dish(name="Burger", description="Burger with fried vegetable patty", price=250, ingredients=[ingredient2, ingredient3])

menu1 = Menu(name="----- Menu -----", dishes=[dish1, dish2, dish3])

restaurant = Restaurant(name="Star Restaurant", menus=[menu1])

# Display restaurant menus
restaurant.display_menus()
```

```python
#Q20. Create a Python class for a social media application, using composition to represent users, posts, and comments.

print("Q20. Create a Python class for a social media application, using composition to represent users, posts, and comments.\nAns.\n")

class Comment:
    def __init__(self, user, content):
        self.user = user
        self.content = content

    def __str__(self):
        return f"{self.user}: {self.content}"

class Post:
    def __init__(self, user, content):
        self.user = user
        self.content = content
        self.comments = []

    def add_comment(self, user, content):
        comment = Comment(user, content)
        self.comments.append(comment)

    def __str__(self):
        post_str = f"{self.user} posts:\n"
        post_str += f"\t{self.content}\n"
        post_str += "\n\tComments:\n"
        for comment in self.comments:
            post_str += f"\t\t{comment}\n"
        return post_str

class User:
    def __init__(self, name):
        self.name = name
        self.posts = []

    def create_post(self, content):
        post = Post(self.name, content)
        self.posts.append(post)
        return post

    def __str__(self):
        return f"User: {self.name}"

class SocialMediaApp:
    def __init__(self, name):
        self.name = name
        self.users = []

    def add_user(self, name):
        user = User(name)
        self.users.append(user)
        return user

    def display_timeline(self, user):
        print(f"{self.name}\n")
        for post in user.posts:
            print(post)
            print()

## Driver code
app = SocialMediaApp(name="!!!BirdChirp App!!!")

user1 = app.add_user("Prince")
user2 = app.add_user("Saina")

post1 = user1.create_post("Hello, everyone! Today is a great day..")
post2 = user2.create_post("Good morning!")

post1.add_comment(user2.name, content="Hi, Prince!")
post2.add_comment(user1.name, content="Hello, Saina..")

app.display_timeline(user1)
app.display_timeline(user2)
```
