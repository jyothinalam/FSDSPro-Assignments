```python
#Q1 Find the value of T(2) for the recurrence relation T(n) = 3T(n-1) + 12n, given that T(0)=5.

print("Q1. Find the value of T(2) for the recurrence relation T(n) = 3T(n-1) + 12n, given that T(0)=5.\nAns.")

print('''
T(0) = 5
T(n) = 3T(n-1) + 12n

n = 1, Substitute n value in recurrence relation. 
T(1) = 3T(1-1) + 12 * 1
     = 3T(0) + 12
     = 3 * 5 + 12  ---------- given T(0) = 5
     = 15 + 12 = 27
T(1) = 27
    
n = 2, Substitute n value in recurrence relation. 
T(2) = 3T(2-1) + 12 * 2
     = 3T(1) + 24
     = 3 * 27 + 24 ----- substitute T(1) = 27
     = 81 + 24 = 105
T(2) = 105

T(0) = 5
T(1) = 27
T(2) = 105

Value of T(2) = 105''')
```

```python
#Q3. Given a recurrence relation, solve it using the recursive tree approach:

print("Q3. Given a recurrence relation, solve it using the recursive tree approach.\nAns.\n")

print("a. T(n) = 2T(n-1) + 1")

print('''

T(n) = 2T(n-1) + 1

                                        1    ----------- 1
                                       /  \\
                                      /    \\
                                     /      \\
                                    n-1     n-1 --------- 2
                                    / \\     / \\
                                   /   \\   /   \\
                                  /     \\ /     \\
                                 n-2  n-2 n-2  n-2 ---------4
                                /   \\
                               /     \\
                              /       \\
                            n-3      n-3 . . . . .  --------- 8
                                .
                                .
                                .
                        /
                       /
                      /
                     T(0) ......................... --------- k   
                    
                    
         T(n) = 1+2+4+8+......+k
              = 1+2^1+2^2+2^3+......+2^k ----------- This is in GP series
            
        Sum of GP series = a(r^k-1)/r-1 = 1(2^k-1)/2-1 = 2^k-1
                         = 2^n = O(2^n)   
        

Recurrence relation: T(n) = 2T(n-1) + 1
Time Complexity    : O(2^n)

-----------------------------------------------------------------------------------\n''')

print("b. T(n) = 2T(n/2) + n")

print('''

T(n) = 2T(n/2) + n
                                       n   ------------ n
                                      /  \\
                                     /    \\
                                    /      \\
                                   /        \\
                                  /          \\
                                 /            \\
                                n/2           n/2 --------n
                               / \\            / \\   
                              /   \\          /   \\
                             /     \\        /     \\
                            n/2^2  n/2^2  n/2^2  n/2^2 -----n   
                            .
                            .
                            .
                            .
                        n/2^k .....................n/2^k ----- n
                        
                T(n) = n + n + n .....+ n = n.k
                     
        Substituting n/2^k = 1 

        => n = 2^k
        => logn = k log2  ------ log2 = 1
        => logn = k

    Substituting for k, we get

    T(n) = n.k
         = n.logn
         = O(nlogn)

Recurrence relation: T(n) = 2T(n/2) + n
Time Complexity    : O(nlogn)
                            
''')
```
